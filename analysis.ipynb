```python
with open("./seeding_optim.pkl", "rb") as doc:
    optim_x = load(doc)

with open("./seeding_sample.pkl", "rb") as doc:
    samp_x = load(doc)
```

```python
print(optim_x[-1])
```

```python
d = {}
for _x in samp_x:
    if _x._score not in d:
        d[_x._score] = [_x]
    else:
        d[_x._score].append(_x)
del d[None]
print(max(d.keys()))
#print(d[max(d.keys())])
print(d[max(d.keys())][0])
#print(d[max(d.keys())][0].mlb)
```

```python
seed_by_team = {k.id: [] for k in sorted(samp_x[0].teams, key=lambda x: x.seed)}
for samp in samp_x:
    for n, t in enumerate(samp.teams):
        seed_by_team[t.id].append(n)
```

```python
df = pd.DataFrame.from_dict(seed_by_team)
variances = pd.DataFrame(df.var())
skews = pd.DataFrame(df.skew())
covariances = df.cov() * -(1 - np.identity(64))
```

```python
# Variance and Performance
X = np.concatenate((pfdf[variances.index.values.tolist()].values.T, pfdf[variances.index.values.tolist()].values.T**2, np.ones_like(variances.values)), axis=-1)
y = variances.values
m = sm.OLS(y, X)
print(m.fit().summary())
plt.scatter(X[:, 0], y)
plt.show()
```
